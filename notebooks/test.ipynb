```python
import cv2
cam = cv2.VideoCapture(c.RAW_VIDEO_PATH)  #type:ignore
print(cam.isOpened())
```

```python
model = YOLO(c.YOLO11N_PATH, task = "detect", verbose = False)
```

```python
def mouse_callback(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"x: {x}, y: {y}")
```

```python
ff = 10
counter = ff
while True:
    counter -= 1
    if counter == -1:
        counter = ff
    print(counter)
    if counter != ff:
        continue
    ret, frame = cam.read()
    if not ret:
        print("Can't get frame")
        break
    else:
        print("Can get frame")

    results = model(frame,classes = [0], imgsz = c.HALF_IMAGE_SIZE)
    result = list(results)[0]
    
    annotated_frame = result.plot()
    wintitle = "Detect person"
    cv2.imshow(wintitle, annotated_frame)

    cv2.waitKey(1)
    
    
    if cv2.getWindowProperty(wintitle, cv2.WND_PROP_VISIBLE) < 1:
        break

cam.release()
cv2.destroyAllWindows()
```

Clicker to get the coordinates

```python
import time, numpy as np

ret, raw_frame = cam.read()

h, w = raw_frame.shape[:2]
f = w
mtx = np.array([
    [f, 0, w/2],
    [0, f, h/2],
    [0, 0, 1]
])
dist = np.array([-0.35, 0.05, 0.0, 0.0, 0.0], dtype = np.float32).reshape(1,5)
frame = cv2.undistort(raw_frame, mtx, dist)

time.sleep(2)
if not ret:
    print("can't get frame")
else: 
    print("could get frame")

WINDOW_TITLE = "Zone Detector"
cv2.namedWindow(WINDOW_TITLE)
cv2.setMouseCallback(WINDOW_TITLE, mouse_callback)

cv2.imshow(WINDOW_TITLE,frame)
cv2.waitKey(0)



cam.release()
```
